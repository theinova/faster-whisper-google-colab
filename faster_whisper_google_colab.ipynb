```python
import os
import glob

from faster_whisper import WhisperModel

model_size = "large-v2"
# Run on GPU with FP16
model = WhisperModel(model_size, device="cuda", compute_type="float16")

# or run on GPU with INT8
#model = WhisperModel(model_size, device="cuda", compute_type="int8_float16")
# or run on CPU with INT8
#model = WhisperModel(model_size, device="cpu", compute_type="int8")

# Path to the top-level folder
top_folder = "/content/drive/MyDrive/audio"

# Find all MP3 files recursively within the top-level folder
mp3_files = glob.glob(os.path.join(top_folder, '**/*.mp3'), recursive=True)

# Loop over all MP3 files and transcribe them using the model
for mp3_file in mp3_files:
    segments, info = model.transcribe(mp3_file, beam_size=5)

    # Construct the output file path
    relative_path = os.path.relpath(mp3_file, top_folder)
    output_file = os.path.join(top_folder, "output", relative_path + ".txt")

    # Create the output directory if it doesn't exist
    os.makedirs(os.path.dirname(output_file), exist_ok=True)

    # Write the transcription results to the output file
    with open(output_file, 'w') as f:
        f.write("Detected language '%s' with probability %f\n" % (info.language, info.language_probability))
        for segment in segments:
            f.write("[%.2fs -> %.2fs] %s\n" % (segment.start, segment.end, segment.text))
```
